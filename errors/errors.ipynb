```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
from datetime import datetime
pd.set_option('display.width', 1000)
import requests
```

```python
errs = pd.read_csv('errors.txt', delimiter='|', names=['simerr', 'count'])
errs.head(n=10)
```

```python
def errcleaner(err):
    for errStr in ['inaccurate payment', 'nonce too low', 'invalid gas used', 'gas limit reached',
                   'invalid merkle root', 'missing trie node', 'nonce too high']:
        if errStr in err:
            return errStr
    return err
errs['simerr'] = errs.apply(lambda row: errcleaner(row['simerr']),axis=1)
errs 
```

```python
errs = errs.groupby('simerr').agg('sum').reset_index()
errs = errs.sort_values('count', ascending=False).reset_index(drop=True)
pd.set_option('max_colwidth', 800)
errs
```
